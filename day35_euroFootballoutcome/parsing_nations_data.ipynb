```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup

def extract_germany_table(soup):
    # Iterate through each h3 to find the one that includes the Germany text
    for h3 in soup.find_all('h3'):
        # Check if 'Germany' is in any of the child elements' strings
        if 'Germany' in h3.get_text():
            # Find the next table sibling
            table = h3.find_next("table", class_="wikitable")
            if table:
                return table
    return None  # Return None if no table found

def parse_table(table):
    headers = [th.get_text(strip=True) for th in table.find_all('th')]
    rows = table.find_all('tr')[1:]  # skip the header row
    data = []
    for row in rows:
        cols = [td.get_text(strip=True) for td in row.find_all('td')]
        data.append(cols)
    return pd.DataFrame(data, columns=headers)

# URL of the UEFA Euro 2024 squads
url = "https://en.wikipedia.org/wiki/UEFA_Euro_2024_squads"
soup = fetch_page(url)

# Extract the table for Germany
germany_table = extract_germany_table(soup)

if germany_table:
    # Convert the table to a DataFrame
    germany_df = parse_table(germany_table)
    print(germany_df)
else:
    print("Germany team table not found.")
```
